```python
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
data["Survived"].value_counts()
```

```python
import seaborn as sns
sns.countplot(x=data['Survived'],hue=data['Pclass'])
plt.title("Count of Survival categorized with Passenger Class ")
plt.show()
```

```python
sns.countplot(data=data, x="Sex", hue='Survived')
plt.title("Orange color representing no. of male and female are survived ")
plt.show()
```

**USING GRAPHS FOR BETTER VISUALIZATION OF THE DATASET**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Distribution of Passengers\' Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='Pclass', y='Survived', data=data, color='yellow')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='SibSp', y='Survived', data=data)
plt.title('Survival Rate by Number of Siblings/Spouses Aboard')
plt.xlabel('Number of Siblings/Spouses Aboard')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=data, color='red')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Gender')
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', hue='Survived', data=data, palette='autumn')
plt.title('Age Distribution by Passenger Class and Survival')
plt.show()
```
